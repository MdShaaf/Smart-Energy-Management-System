```python
import pandas as pd
import ast

# Assuming you've already loaded the data into df
# df = pd.read_csv('your_file.csv', sep='\t')

# Function to parse the array string
df = df_solar.copy()
import pandas as pd
import numpy as np
import re

# Function to parse the array string
def parse_array_column(array_str):
    # Remove outer quotes
    array_str = array_str.strip('"')
    
    # Extract all the timestamp-value pairs using regex
    pattern = r"array\(\['([^']+)',\s*'([^']+)'\]"
    matches = re.findall(pattern, array_str)
    
    return matches

# Parse the data column
df['parsed_data'] = df['data'].apply(parse_array_column)

# Explode to create one row per timestamp-value pair
df_exploded = df.explode('parsed_data')

# Extract timestamp and value from each tuple
df_exploded['timestamp'] = df_exploded['parsed_data'].apply(lambda x: x[0] if isinstance(x, tuple) else None)
df_exploded['value'] = df_exploded['parsed_data'].apply(lambda x: x[1] if isinstance(x, tuple) else None)

# Convert to appropriate types
df_exploded['value'] = pd.to_numeric(df_exploded['value'])
df_exploded['timestamp'] = pd.to_datetime(df_exploded['timestamp'], format='mixed')

# Drop the temporary columns
df_exploded = df_exploded.drop(['data', 'parsed_data'], axis=1)

# Reset index
df_exploded = df_exploded.reset_index(drop=True)

print(df_exploded.head(20))
print(f"\nTotal rows: {len(df_exploded)}")
```
