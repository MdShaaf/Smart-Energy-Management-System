```python
import pandas as pd
import numpy as np

# A simulated chaotic dataset to practice string cleaning and formatting
raw_data = {
    'raw_timestamp': [
        "2025-12-31T19:00",      # ISO Standard
        "2025/12/31 19:15:00",   # Slashes and spaces
        "31-12-2025 19:30",      # European style
        "2025-12-31T19:45:30Z",  # UTC format
        None                     # Missing entry
    ],
    'temperature_dirty': [
        "23.5 °C",               # Standard string with units
        "  24.1 C ",             # Extra leading/trailing spaces
        "22.8#",                 # Corrupted symbol
        25.3,                    # Already a float! (This throws the error)
        "24.0"                   # Clean string decimal
    ],
    'energy_demand_kw': [
        "4,500 kw",              # Commas as thousands separators
        "4650   ",               # Massive trailing whitespace
        "4-700",                 # Typo hyphen instead of a number separator
        "4750kw",                # No spaces between number and text
        "UNKNOWN"                # Text anomaly
    ]
}

df = pd.DataFrame(raw_data)
print("--- RAW DILEMMAS DATAFRAME ---")
print(df)
print("\n--- COLUMN DATA TYPES ---")
print(df.dtypes)
```
